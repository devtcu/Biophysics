Each cell is it's own individual fitting of a particular data set gathered from to two different graph. One measures the expression of HXB2 and the other gp120/gp41 expression. These blocks of code are used to return five main things
* SSR(Sum of Square Residual)
* Gamma(Syncytia Formation Rate) for that level of dosage
* K-value(Reciprocal of this value of the fusing rate in two step fusion) for that level of dosage
* A chart showing the fitting data against the true experimental data
* $\chi^2$ for each one

Everything  is working as it should, as well as Akaike's Information Criterion has been implemented.

Once these five things are obtained, they are stored in a respective variable. Once these variables are all obatined then they are spliced together in an array which the amount of dosage that was used to achieved these results. These results then are plotted in a dose-response chart.
